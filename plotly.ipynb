```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly as py
import plotly.graph_objects as go
import plotly.express as px
```

1. Write a program to draw a scatter plot for random 1000 x and y coordinates

```python
ran_x = np.random.rand(1000)
ran_y = np.random.rand(1000)

fig = go.Figure(data=go.Scatter(x=ran_x, y=ran_y, mode='markers'))
fig.show()
```

```python
import plotly.express as px
import numpy as np 

def draw_scatter_plot():    
    """
       Description:
          Function is used to draw scatter plot of random number using plotly library.
       Parameter:
          No parameter required.
       Return:
          Returns nothing.
    """

    # creating random data through randomint 
    # function of numpy.random 

    n = 1000
    random_x = np.random.randn(n)
    random_y = np.random.randn(n)

        

    plot = px.scatter(random_x, random_y, color_discrete_sequence=px.colors.qualitative.Antique)
    #plot = px.line(random_x, random_y, color_discrete_sequence=px.colors.qualitative.Antique)

    plot.show()

if __name__ == "__main__":
    draw_scatter_plot()
```

2. Write a program to draw line and scatter plots for random 100 x and y coordinates

3. Write a program to draw a scatter plot for random 500 x and y coordinates and style it

```python
ran_x = np.random.rand(500)
ran_y = np.random.rand(500)

fig = go.Figure(data=go.Scatter(x=ran_x, y=ran_y, mode='markers',marker=dict(color='green', size=8, line=dict(color='black', width=1))))


fig.show()
```
