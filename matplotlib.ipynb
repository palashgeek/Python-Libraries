3. Write a Python program to draw a line using given axis values taken from a text file, with
suitable label in the x axis, y axis and a title.
Test Data:
test.txt
1 2
2 4
3 1

```python
x = []
y = []
for line in open('test.txt', 'r'):
    lines = [i for i in line.split()]
    x.append(lines[0])
    y.append(int(lines[1]))

plt.plot(x,y, linestyle='-')
plt.title("Test Data",color = 'red')
plt.xlabel('Index')
plt.ylabel('Data')
plt.show()
```

5. Write a Python program to plot two or more lines on same plot with suitable legends of each
line.

```python
x1 = [10,20,30]
y1 = [20,40,10]
plt.plot(x1,y1, color = 'blue', linestyle='-') 

x2 = [10,20,30]
y2 = [40,10,30]
plt.plot(x2,y2)
plt.plot(x1,y1, color = 'pink', linestyle='-.') 
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("Lines")
plt.legend(["Blue","Green"])
```

7. Write a Python program to plot two or more lines with different styles

```python
x = [1,2,3,4,5]
y = [11,22,33,44,55]
x1 = [1,2,3,4,5]
y2 = [33,44,55,11,22]
plt.xlabel('Count')
plt.ylabel('eleven')
plt.plot(x,y, color ='pink', linewidth = 3, label = 'line1-dotted',linestyle ='--')
plt.plot(x1,y2, color ='orange', linewidth = 2, label = 'line2-dashed',linestyle ='-')
plt.legend()
plt.show()
```

8. Write a Python program to plot two or more lines and set the line markers.

markers == representation of a point on a graph/chart

```python
x = [1,2,3,4,5]
y = [11,22,33,44,55]
x1 = [1,2,3,4,5]
y2 = [33,44,55,11,22]
plt.xlabel('Count')
plt.ylabel('eleven')
plt.plot(x,y, color ='pink', linewidth = 3, label = 'line1-dotted',linestyle ='--', marker = "D", mec="r")
plt.plot(x1,y2, color ='orange', linewidth = 2, label = 'line2-dashed',linestyle ='-', marker = "o", mec="b")
plt.legend()
plt.show()
```

9. Write a Python program to display the current axis limits values and set new axis values.

```python
x = [1,0,2,3,4]
y = [1,2,3,1,2]
plt.plot(x,y, c="black", marker="o", mfc="blue", ms= 20)
plt.xlabel("X")
plt.ylabel("Y")

plt.xlim([-0.5,4.5])
plt.ylim([0.5,3.5])
plt.show()
```
