Seaborn Scatterplot

1. Write a program to draw a scatter plot of day against total bill for a dataset given in a
url
https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv

```python
tips_df = sns.load_dataset('tips')
tips_df.head()
```

```python
sns.scatterplot(x='day', y='total_bill',style = 'sex',hue='smoker',data = tips_df)
plt.title("A Restro Bill")
```
